```python
##Preprocesing and cleaning the data
##Positive review
data['rating']=data['rating'].apply(lambda x: 1 if x>3 else 0)
```

```python
data['rating'].unique()
data['rating'].value_counts()
```

```python
##1. Lower All the cases
data['reviewText'].str.lower()
```

```python
data.head()
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
## Removing special characters
data['reviewText']=data['reviewText'].str.replace('[^a-z A-Z]+',     ' ',regex=True) 
 
## remove  the stopwords
data['reviewText']=data['reviewText'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopwords.words('english'))]))
##remove url
data['reviewText']=data['reviewText'].str.replace('http\S+|www.\S+','',case=False)
## remove html tags
data['reviewText']=data['reviewText'].str.replace('<.*?>','',regex=True)
## remove punctuation       
data['reviewText']=data['reviewText'].str.replace('[^\w\s]','',regex=True)
```

```python
data.head()
```

```python
##Word lemmatization
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatizer_words(text):
    return ' '.join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
data['reviewText']=data['reviewText'].apply(lambda x:lemmatizer_words(x))
```

```python
##Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data['reviewText'],data['rating'],test_size=0.2,random_state=42)
```

```python
##BOW Word2Vec
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
bow = CountVectorizer()
X_train_bow=bow.fit_transform(X_train).toarray()
X_test_bow=bow.transform(X_test).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
tfidf = CountVectorizer()
X_train_tfidf=bow.fit_transform(X_train).toarray()
X_test_tfidf=bow.transform(X_test).toarray()
```

```python
X_train_bow
```

```python
from sklearn.naive_bayes import GaussianNB
nb_model_bow=GaussianNB().fit(X_train_bow,y_train)
nb_model_tfidf=GaussianNB().fit(X_train_tfidf,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
y_pred_bow=nb_model_bow.predict(X_test_bow)
```

```python
y_pred_tfidf=nb_model_tfidf.predict(X_test_tfidf)
```

```python
confusion_matrix(y_test,y_pred_bow)
```

```python
print("BOW accuracy: ",accuracy_score(y_test,y_pred_bow))
```

```python
print("TF-IDF accuracy: ",accuracy_score(y_test,y_pred_tfidf))
```
